```python
## Start pandas

import pandas as pd

## import dataframe csv

df = pd.read_csv('LIF Ashland Nit 12-13-18.csv')

## rename columns for easy manipulation

df.rename(columns = {'Sample Name':'ID', '410 nm   [A]':'B'}, inplace=True)

## group csv by desired column

g = df.groupby('ID')

## adding column to original dataframe with means for each group based on sample id, and rename column

final=df.merge(g.mean(),on='ID')
final2=final.merge(abs(g.max()-g.min())/g.max(), on='ID')
final2.rename(columns = {'B':'Var', 'B_x':'Abs', 'B_y':'Mean'}, inplace=True)
final2
```

```python
## testing to find how to call values of specific cells

for key, row in final2.iterrows():
   print(key, row['ID'], row['Var'])

## checking type to see if it will be compatible for comparison functions

print(type(row['Var']))

## float, so good to run conditional:
## for each sample, if the value in the Var column is greater than 0.1, 
## display an error message... if it is less than 0.1, display the value 
## from that sample's Mean column.

for key, row in final2.iterrows():
    if float(row['Var']) > 0.1:
        print(key, row['ID'], "Error")
    else:
        print(key, row['ID'], row['Mean'])

for key, row in final2.iterrows():
    if float(row['Var']) > 0.1:
        print("Error")
    else:
        print(row['Mean'])
        
## Now that I know how to call this, I'll change the function from print to
## creating a new columns in my original dataframe file. This can be accomplished
## by creating a LIST of if/else values based on row, and then adding them in a
## new column to the dataframe.

Results = []

for key, row in final2.iterrows():
    if float(row['Var']) > 0.1:
        Results.append('Error')
    else:
        Results.append(row['Mean'])

final2['Results'] = Results
final2

## this will give a new pandas dataframe with the "Results" column displaying either the mean
## of the samples or an error message, indicating whether or not the values are acceptable.

## to print a new dataframe with only the results and id number, use the following code:

complete = final2.drop(final2.columns[[1, 2, 3]], axis=1)
complete
```
